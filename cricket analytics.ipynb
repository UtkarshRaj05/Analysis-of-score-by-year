```python
# Visualizing Top 10 Players by Total Runs
top_players = df.groupby("player_name")["total_runs"].sum().nlargest(10).reset_index()
fig = px.bar(top_players, x="player_name", y="total_runs", title="Top 10 Players by Total Runs", color_continuous_scale="Blues")
fig.show()
```

**Conclusion:**

Virat Kohli is the most prolific run-scorer in this dataset, with a noticeable lead over the rest.

The difference in total runs among the players suggests varying levels of consistency, longevity, and possibly strike rates.

Players like MS Dhoni and AB de Villiers, despite being known for aggressive batting, still maintain high total runs, showing their impact across formats.

The presence of players from different generations highlights a mix of classic and modern cricketers. *italicized text*

```python
ball_face = df.groupby("player_name")["total_balls_faced"].sum().nlargest(10).reset_index()
fig = px.bar(ball_face, x="player_name", y="total_balls_faced", title="balls faced by batsman")
fig.show()
```

**Conclusion:**

Virat Kohli and Kumar Sangakkara have demonstrated high endurance and consistency in facing deliveries, indicating a long and stable batting career.

The drop in the number of balls faced is gradual across players, suggesting a ranking of batting longevity or experience.

AB de Villiers and Michael Clarke, despite being great batsmen, have faced fewer balls compared to the others in the dataset *italicized text*

```python
fig = px.bar(df, x = "team", y = "total_runs", title = "Total Runs by Team", color = "team", color_discrete_sequence = px.colors.qualitative.Set1)
fig.show()
```

**Key Observations:**

India and Australia have scored the highest total runs, surpassing 100,000 runs, indicating their strong and consistent batting performances over time.

Sri Lanka, South Africa, and England follow closely, all having scored between 80,000 and 100,000 runs, showing their long-standing presence as competitive cricketing nations.

New Zealand, West Indies, Pakistan, and Bangladesh have accumulated a moderate number of runs, indicating their balanced cricketing history.
Teams like Zimbabwe, Ireland, and Scotland have significantly fewer total runs, suggesting fewer matches played or a relatively shorter cricketing history.

Associate nations (such as UAE, Namibia, Papua New Guinea, Netherlands, Nepal, Oman, Kenya, Canada, and the USA) have much lower total runs, reflecting their limited participation in international cricket.

**Conclusion:**

India and Australia dominate in terms of total runs, showing their strong batting performances over time.

Sri Lanka, South Africa, and England are also among the top-scoring teams, reflecting their consistent cricketing history.

Smaller and associate nations have significantly fewer total runs, likely due to playing fewer international matches.

This graph highlights the historical dominance of traditional cricketing powerhouses in terms of total runs scored.

```python
# Select the top 10 players by total runs
top_players = df.groupby("player_name")["total_runs"].sum().nlargest(10).reset_index()

# Filter the dataset for only these top 10 players
filtered_data = df[df["player_name"].isin(top_players["player_name"])]

# Create scatter plot
fig = px.scatter(filtered_data, x = "strike_rate", y = "total_runs", color = "player_name",title = "Strike Rate vs Total Runs (Top 10 Players)", size="total_runs")
fig.show()
```

**Key Observations:**

Virat Kohli (V Kohli) has the highest total runs (~14k) while maintaining a strong strike rate, making him one of the most consistent and impactful batsmen.

AB de Villiers has a high strike rate along with significant total runs, indicating his aggressive and fast-scoring approach.

Rohit Sharma (RG Sharma) and MS Dhoni also have high total runs (~10k-11k) but with a slightly lower strike rate compared to AB de Villiers.

Players like Tamim Iqbal, HM Amla, and DPMD Jayawardene have relatively lower total runs and moderate strike rates, indicating a more steady and traditional batting style rather than explosive scoring.
Most players are clustered around a similar total runs range (~9k-11k), but their strike rates vary significantly, showing different playing styles.

**Conclusion:**

Virat Kohli is the top scorer with a balanced strike rate.
AB de Villiers stands out with a high strike rate, making him a more aggressive batsman.

Players like Tamim Iqbal and HM Amla have lower strike rates, indicating a more stable and anchor-like role in their teams.

This graph highlights the difference between accumulators (high total runs, moderate strike rate) and aggressive strikers (high strike rate, moderate total runs).

```python
# Select the top 10 players by total wickets taken
top_bowlers = df.groupby("player_name")["total_wickets_taken"].sum().nlargest(10).reset_index()

# Create bar graph for top 10 bowlers
fig = px.bar(top_bowlers, x="player_name", y="total_wickets_taken",
             title="Top 10 Players with Most Wickets",
             color="player_name", color_discrete_sequence=px.colors.qualitative.Set2)
fig.show()
```

**Key Insights:**

Lasith Malinga stands out as the most dominant bowler in this dataset.
Pace bowlers dominate the list, with only a couple of spinners like Imran Tahir, Kuldeep Yadav, and R Ashwin making it.

Malinga’s lead suggests his long and successful career in taking wickets consistently across matches.

The competition among the rest of the bowlers is relatively close, with all taking between 160-250 wickets.
